````python
x = np.random.rand(10,3)
````

이제 생성된 어레이의 각 항목들과 0.5 사이의 거리를 계산한다.

````python
a= np.abs(x - 0.5)
````

`argsort`를 이용하여 어레이 `x`의 각 행별로 0.5로부터 가장 가까운 값의 위치를 알아낸다.

````python
b = a.argsort()[:,0]
````

위 정보를 이용하여 x의 각 행별로 0.5로부터 가장 가까운 값이 얼마인지를 다음과 같아 알아낼 수 있다.

````python
min05 = x[np.arange(10), b]
````
